```python
import pandas as pd
%matplotlib inline
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()
df['x'] = random.sample(range(1, 100), 25)
df['y'] = random.sample(range(1, 100), 25)
# error: df.show()  #'DataFrame' object has no attribute 'show'
df.head()
```

```python
#Scatterplot
sns.lmplot('x', 'y', data = df, fit_reg=False)
```

```python
#Density Plot
sns.kdeplot(df.y)
```

```python
sns.distplot(df.x)
```

```python
#Histogram
plt.hist(df.x, alpha=.3)
sns.rugplot(df.x);
```

```python
#Boxplot
sns.boxplot([df.y, df.x])
```
